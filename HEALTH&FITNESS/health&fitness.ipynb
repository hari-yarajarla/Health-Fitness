```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import GradientBoostingClassifier

# Define feature types
categorical_features = ['Gender', 'Occupation', 'BMI Category', 'Blood Pressure']
numeric_features = ['Age', 'Sleep Duration', 'Physical Activity Level', 'Stress Level', 'Heart Rate', 'Daily Steps']


# Preprocessor
preprocessor = ColumnTransformer([
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
], remainder='passthrough')

# Full pipeline
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', GradientBoostingClassifier(random_state=42))
])

# Fit pipeline
pipeline.fit(df[categorical_features + numeric_features], df['Quality of Sleep'])

# Save pipeline
import joblib
joblib.dump(pipeline, 'health_pipeline.pkl')
```

```python
import gradio as gr
import pandas as pd
import joblib
import numpy as np

# Load the trained pipeline
pipeline = joblib.load('health_pipeline.pkl')

# Define placeholder values based on the training data (from df)
# Using mode for categorical and mean for numerical features from the original dataframe
placeholder_values = {}
for col in ['Occupation', 'BMI Category', 'Blood Pressure']:
    placeholder_values[col] = df[col].mode()[0]
for col in ['Heart Rate', 'Daily Steps']:
     placeholder_values[col] = df[col].mean()


def predict_sleep_quality(gender, age, physical_activity, stress_level, sleep_duration):
    # Create a DataFrame with the input features and placeholder values for missing features
    input_data = {
        'Gender': [gender],
        'Age': [age],
        'Sleep Duration': [sleep_duration],
        'Physical Activity Level': [physical_activity],
        'Stress Level': [stress_level],
        'Occupation': [placeholder_values['Occupation']],
        'BMI Category': [placeholder_values['BMI Category']],
        'Blood Pressure': [placeholder_values['Blood Pressure']],
        'Heart Rate': [placeholder_values['Heart Rate']],
        'Daily Steps': [placeholder_values['Daily Steps']]
    }

    input_df = pd.DataFrame(input_data)

    # Ensure the order of columns matches the pipeline's expected order
    # The pipeline expects the columns in the order they were presented during training
    # This order is derived from categorical_features + numerical_features when the preprocessor was defined
    expected_columns_order = ['Gender', 'Occupation', 'BMI Category', 'Blood Pressure', 'Age', 'Sleep Duration', 'Physical Activity Level', 'Stress Level', 'Heart Rate', 'Daily Steps']
    input_df = input_df[expected_columns_order]


    # Make prediction using the pipeline
    prediction = pipeline.predict(input_df)[0]

    return f"Predicted Quality of Sleep: {prediction}"

# Build Gradio UI
iface = gr.Interface(
    fn=predict_sleep_quality,
    inputs=[
        gr.Radio(["Male", "Female"], label="Gender"),
        gr.Number(label="Age"),
        gr.Number(label="Physical Activity Level"),
        gr.Number(label="Stress Level"),
        gr.Number(label="Sleep Duration")
    ],
    outputs=gr.Text(label="Predicted Sleep Quality"),
    title="Sleep Quality Predictor",
    description="Predicts sleep quality based on selected inputs."
)

iface.launch(debug=True)
```
